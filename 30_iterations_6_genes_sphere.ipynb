```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
```

```python
# 1st hyperparameter settings
sphere_pop100_mutrate0_1_mutrange0_1 = glob.glob('csv_results/30_iterations_6_genes/sphere_experiment_pop100_mutrate0.1_mutrange0.1/fitness_data/*.csv')
csv_list = []

for csv_file in sphere_pop100_mutrate0_1_mutrange0_1:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop100_mutrate0_1_mutrange0_1_df = pd.concat(csv_list, ignore_index=True)

# sphere_pop100_mutrate0_1_mutrange0_1_df.head(5)
sphere_pop100_mutrate0_1_mutrange0_1_df.max()
```

```python
# 2nd hyperparameter settings
sphere_pop100_mutrate0_1_mutrange0_4 = glob.glob('csv_results/30_iterations_6_genes/sphere_experiment_pop100_mutrate0.1_mutrange0.4/fitness_data/*.csv')
csv_list = []

for csv_file in sphere_pop100_mutrate0_1_mutrange0_4:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop100_mutrate0_1_mutrange0_4_df = pd.concat(csv_list, ignore_index=True)

# sphere_pop100_mutrate0_1_mutrange0_4_df.head(5)
sphere_pop100_mutrate0_1_mutrange0_4_df.max()
```

```python
# 3rd hyperparameter settings
sphere_pop100_mutrate0_1_mutrange0_8 = glob.glob('csv_results/30_iterations_6_genes/sphere_experiment_pop100_mutrate0.1_mutrange0.8/fitness_data/*.csv')
csv_list = []

for csv_file in sphere_pop100_mutrate0_1_mutrange0_8:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop100_mutrate0_1_mutrange0_8_df = pd.concat(csv_list, ignore_index=True)

# sphere_pop100_mutrate0_1_mutrange0_8_df.head(5)
print(sphere_pop100_mutrate0_1_mutrange0_8_df.max())
print(sphere_pop100_mutrate0_1_mutrange0_8_df["max_distance"].idxmax())
```

```python
# 4th hyperparameter settings
sphere_pop100_mutrate0_4_mutrange0_4 = glob.glob('csv_results/30_iterations_6_genes/sphere_experiment_pop100_mutrate0.4_mutrange0.4/fitness_data/*.csv')
csv_list = []

for csv_file in sphere_pop100_mutrate0_4_mutrange0_4:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop100_mutrate0_4_mutrange0_4_df = pd.concat(csv_list, ignore_index=True)

# sphere_pop100_mutrate0_4_mutrange0_4_df.head(5)
sphere_pop100_mutrate0_4_mutrange0_4_df.max()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7))

# define your dataframes
dataframes = [
    sphere_pop100_mutrate0_1_mutrange0_1_df,
    sphere_pop100_mutrate0_1_mutrange0_4_df,
    sphere_pop100_mutrate0_1_mutrange0_8_df,
    sphere_pop100_mutrate0_4_mutrange0_4_df
]

colors = ['blue', 'red', 'green', 'purple', 'orange']
labels = ['100_0.1_0.1', '100_0.1_0.4', '100_0.1_0.8', '100_0.4_0.4']

for dataframe, color, label in zip(dataframes, colors, labels):
    for (i, j), ax in np.ndenumerate(axs):
        metric = ['max_distance', 'avg_distance', 'avg_speed', 'path_straightness'][i*2 + j]

        # plot data
        ax.plot(dataframe.index, dataframe[metric], color=color, label=label)

        # calculate best fit line
        coef = np.polyfit(dataframe.index, dataframe[metric], 1)
        poly1d_fn = np.poly1d(coef)

        # plot best fit line
        ax.plot(dataframe.index, poly1d_fn(dataframe.index), color=color, linestyle='--')

# add legends
for ax in axs.flat:
    ax.legend()

# Set titles
axs[0, 0].set_title('Max Distance (box, 10i, 3g)')
axs[0, 1].set_title('Avg Distance (box, 10i, 3g)')
axs[1, 0].set_title('Avg Speed (box, 10i, 3g)')
axs[1, 1].set_title('Path Straightness (box, 10i, 3g)')

plt.tight_layout()
plt.show()
```
