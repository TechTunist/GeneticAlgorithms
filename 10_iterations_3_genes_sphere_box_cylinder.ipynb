```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
```

### Create separate dataframes for each change in hyperparameter

### These simulations test the effect of population size, mutation rate and mutation range over creatures that evolve using crossover and point mutate, with 50 generations of creatures each having 3 genes

```python
# 1st hyperparameter settings
box_experiment_pop100_mutrate0_1_mutrange0_5 = glob.glob('csv_results/50_iterations_3_genes/box_experiment_pop100_mutrate0.1_mutrange0.5/*.csv')
dfs = []

for csv_file in box_experiment_pop100_mutrate0_1_mutrange0_5:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
box_experiment_pop100_mutrate0_1_mutrange0_5_df = pd.concat(dfs, ignore_index=True)

# box_experiment_pop100_mutrate0_1_mutrange0_5_df.head(5)
```

```python
# 2nd hyperparameter settings
box_experiment_pop100_mutrate0_2_mutrange0_5 = glob.glob('csv_results/50_iterations_3_genes/box_experiment_pop100_mutrate0.2_mutrange0.5/*.csv')
dfs = []

for csv_file in box_experiment_pop100_mutrate0_2_mutrange0_5:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
box_experiment_pop100_mutrate0_2_mutrange0_5_df = pd.concat(dfs, ignore_index=True)

# box_experiment_pop100_mutrate0_2_mutrange0_5_df.head(5)
```

```python
# 3rd hyperparameter settings
box_experiment_pop200_mutrate0_1_mutrange0_5 = glob.glob('csv_results/50_iterations_3_genes/box_experiment_pop200_mutrate0.1_mutrange0.5/*.csv')
dfs = []

for csv_file in box_experiment_pop200_mutrate0_1_mutrange0_5:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
box_experiment_pop200_mutrate0_1_mutrange0_5_df = pd.concat(dfs, ignore_index=True)

box_experiment_pop200_mutrate0_1_mutrange0_5_df.head(5)
```

```python
# 4th hyperparameter settings
box_experiment_pop200_mutrate0_5_mutrange0_5 = glob.glob('csv_results/50_iterations_3_genes/box_experiment_pop200_mutrate0.5_mutrange0.5/*.csv')
dfs = []

for csv_file in box_experiment_pop200_mutrate0_5_mutrange0_5:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
box_experiment_pop200_mutrate0_5_mutrange0_5_df = pd.concat(dfs, ignore_index=True)

box_experiment_pop200_mutrate0_5_mutrange0_5_df.head(5)
```

```python
# 5th hyperparameter settings
box_experiment_pop500_mutrate0_5_mutrange0_5 = glob.glob('csv_results/50_iterations_3_genes/box_experiment_pop500_mutrate0.5_mutrange0.5/*.csv')
dfs = []

for csv_file in box_experiment_pop500_mutrate0_5_mutrange0_5:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
box_experiment_pop500_mutrate0_5_mutrange0_5_df = pd.concat(dfs, ignore_index=True)

# box_experiment_pop500_mutrate0_5_mutrange0_5_df.head(5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7))

# define your dataframes
dataframes = [
    box_experiment_pop100_mutrate0_1_mutrange0_5_df,
    box_experiment_pop100_mutrate0_2_mutrange0_5_df,
#     box_experiment_pop200_mutrate0_1_mutrange0_5_df,
    box_experiment_pop200_mutrate0_5_mutrange0_5_df,
    box_experiment_pop500_mutrate0_5_mutrange0_5_df
]

colors = ['blue', 'red', 'green', 'purple', 'orange']
labels = ['df1', 'df2', 'df3', 'df4', 'df5']

for dataframe, color, label in zip(dataframes, colors, labels):
    for (i, j), ax in np.ndenumerate(axs):
        metric = ['max_distance', 'avg_distance', 'avg_speed', 'path_straightness'][i*2 + j]

        # plot data
        ax.plot(dataframe.index, dataframe[metric], color=color, label=label)

        # calculate best fit line
        coef = np.polyfit(dataframe.index, dataframe[metric], 1)
        poly1d_fn = np.poly1d(coef)

        # plot best fit line
        ax.plot(dataframe.index, poly1d_fn(dataframe.index), color=color, linestyle='--')

# add legends
for ax in axs.flat:
    ax.legend()

# Set titles
axs[0, 0].set_title('Max Distance (box, 50i, 3g)')
axs[0, 1].set_title('Avg Distance (box, 50i, 3g)')
axs[1, 0].set_title('Avg Speed (box, 50i, 3g)')
axs[1, 1].set_title('Path Straightness (box, 50i, 3g)')

plt.tight_layout()
plt.show()
```

The graphs above show a massive outlier, perhaps due to random favourable conditions from the outset of gene creation. The steepest line of best fit of all the simulations will be used to indicate favourable hyperparameters
