### Since spheres gave the best result, lets explore sphere creatures with further hyperparemeter analysis

```python
# 1st hyperparameter settings
sphere_pop100_mutrate0_1_mutrange0_5 = glob.glob('csv_results/10_iterations_6_genes_grow_mutate_added/sphere_experiment_pop100_mutrate0.1_mutrange0.5/*.csv')
csv_list = []

for csv_file in sphere_pop100_mutrate0_1_mutrange0_5:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop100_mutrate0_1_mutrange0_5_df = pd.concat(csv_list, ignore_index=True)

sphere_pop100_mutrate0_1_mutrange0_5_df.head(5)
```

```python
# 3rd hyperparameter settings
sphere_pop200_mutrate0_1_mutrange0_5 = glob.glob('csv_results/10_iterations_6_genes_grow_mutate_added/sphere_experiment_pop200_mutrate0.1_mutrange0.5/*.csv')
csv_list = []

for csv_file in sphere_pop200_mutrate0_1_mutrange0_5:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop200_mutrate0_1_mutrange0_5_df = pd.concat(csv_list, ignore_index=True)

sphere_pop200_mutrate0_1_mutrange0_5_df.head(5)
```

```python
# 4th hyperparameter settings
sphere_pop200_mutrate0_5_mutrange0_5 = glob.glob('csv_results/10_iterations_6_genes_grow_mutate_added/sphere_experiment_pop200_mutrate0.5_mutrange0.5/*.csv')
csv_list = []

for csv_file in sphere_pop200_mutrate0_5_mutrange0_5:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop200_mutrate0_5_mutrange0_5_df = pd.concat(csv_list, ignore_index=True)

sphere_pop200_mutrate0_5_mutrange0_5_df.head(5)
```

```python
# 5th hyperparameter settings
sphere_pop500_mutrate0_5_mutrange0_5 = glob.glob('csv_results/10_iterations_6_genes_grow_mutate_added/sphere_experiment_pop500_mutrate0.5_mutrange0.5/*.csv')
csv_list = []

for csv_file in sphere_pop500_mutrate0_5_mutrange0_5:
    df = pd.read_csv(csv_file)
    csv_list.append(df)
    
sphere_pop500_mutrate0_5_mutrange0_5_df = pd.concat(csv_list, ignore_index=True)

sphere_pop500_mutrate0_5_mutrange0_5_df.head(5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7))

# define your dataframes
dataframes = [
    sphere_pop100_mutrate0_1_mutrange0_5_df,
    sphere_pop100_mutrate0_2_mutrange0_5_df,
    sphere_pop200_mutrate0_1_mutrange0_5_df,
    sphere_pop200_mutrate0_5_mutrange0_5_df,
    sphere_pop500_mutrate0_5_mutrange0_5_df
]

colors = ['blue', 'red', 'green', 'purple', 'orange']
labels = ['100_0.1_0.5', '100_0.2_0.5', '200_0.1_0.5', '200_0.5_0.5', '500_0.5_0.5']

for dataframe, color, label in zip(dataframes, colors, labels):
    for (i, j), ax in np.ndenumerate(axs):
        metric = ['max_distance', 'avg_distance', 'avg_speed', 'path_straightness'][i*2 + j]

        # plot data
        ax.plot(dataframe.index, dataframe[metric], color=color, label=label)

        # calculate best fit line
        coef = np.polyfit(dataframe.index, dataframe[metric], 1)
        poly1d_fn = np.poly1d(coef)

        # plot best fit line
        ax.plot(dataframe.index, poly1d_fn(dataframe.index), color=color, linestyle='--')

# add legends
for ax in axs.flat:
    ax.legend()

# Set titles
axs[0, 0].set_title('Max Distance (box, 10i, 3g)')
axs[0, 1].set_title('Avg Distance (box, 10i, 3g)')
axs[1, 0].set_title('Avg Speed (box, 10i, 3g)')
axs[1, 1].set_title('Path Straightness (box, 10i, 3g)')

plt.tight_layout()
plt.show()
```
